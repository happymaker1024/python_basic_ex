# <font color='blue'> 2. 각 학생별로, 중간과 기말의 평균점수를 구하세요.  

```python
student_details = [
  {'student_id' : 1, 'subject' : 'math', 'midterm' : 60, 'final' : 85},
  {'student_id' : 2, 'subject' : 'math', 'midterm' : 80, 'final' : 78},
  {'student_id' : 3, 'subject' : 'math', 'midterm' : 90, 'final' : 85}
]
```

```python
# [] 대괄호 시작 대괄호로 끝나는 것을 리스트라고 함.
# {} 중괄호로 시작 중괄호로 끝나는 것을 딕셔너리라고 함.
```

```python
# 각 학생별로, 중간과 기말의 평균점수를 구하세요.
# 한 학생의 중간성적 추출, 기말성적 추출
st1_avg = (student_details[0]['midterm'] + student_details[0]['final']) / 2
```

```python
# 두번째 학생 평균
st2_avg = (student_details[1]['midterm'] + student_details[1]['final']) / 2
```

```python
# 두번째 학생 평균
st3_avg = (student_details[2]['midterm'] + student_details[2]['final']) / 2
```

# <font color='blue'> 3. 위에서 구한 평균점수를,"AVERAGE" 라는 키를 새로 만들어서, 각 학생의 딕셔너리에 추가하세요.

```python
student_details[0]["AVERAGE"] = st1_avg
student_details
```

```python
student_details[1]["AVERAGE"] = st2_avg
student_details[2]["AVERAGE"] = st3_avg
student_details
```

# <font color='blue'> 4. 각각의 숫자를,  제곱한 값으로 바꾸세요.
