```python
my_socre = 50
cut_line = 60
# 참인 경우만 실행 하때
if my_socre > cut_line :
  print("합격")
```

```python
# 참인 경우와 거짓인 경우 각각에 판단해서 실행
if my_socre > cut_line :
  print("합격")
else:
  print("불합격")
```

```python
score = 70
score >= 60 and score < 80
```

```python
# 60 미만 : 더 공부하세요.
# 60~80미만 : 합격 했어요.
# 80~이상 : 최고 예요.
score = 90  # 들어오는 설적
cut_line = 60
if score < cut_line :
  print("더 공부하세요.")
# score >= 60 or  score < 80
elif score >= 60 and score < 80 :   # 또다른 조건
  print("합격 했어요.")
else:
  print("최고 예요.")
```

## Nested 중첩된 조건문

```python
year = 2025
month = 12
day = 31
day = day + 1

if day > 31 :
  month = month + 1
  day = 1

  if month > 12:
    year = year + 1
    month = 1

print(f"{year}년 {month}월 {day}일")
```
