# <font color='blue'> 1. 유저한테서 KPH값을 입력 받으면, 그에 맞는 MPH로 변환하는 코드를 만드세요. </font>
- MPH(Miles per Hour) = 0.6214 * KPH
- KPH(Kilometers per Hour) = 1.6093 * MPH

```python
# 입력 되는 값은 KPH
kph = input("KMH 값을 입력하세요. > ")
```

```python
kph
```

```python
# kph -> mph
# MPH : 0.6214 * KPH
KPH_TO_MPH = 0.6214
mph = int(kph) * KPH_TO_MPH
```

```python
# kph -> mph 변환값
mph
```

```python
# mph -> kph
```

```python
mph * 1.6093
```

## 부동소수점 연산을 정확하게 하려면

```python
# 정확한 변환 계수 사용
MILE_TO_KM = 1.609344  # 정확한 변환 계수
KM_TO_MILE = 1 / MILE_TO_KM
```

```python
KM_TO_MILE
```

```python
mph = int(kph) * KM_TO_MILE
mph
```

```python
mph * MILE_TO_KM
```

### <font color='blue'>3.  유저한테 두개의 float을 입력받아서, 두 수의 곱과 합을, 다음처럼 똑같이 출력하는 코드를 만드세요.</font>
## 출력형태 : First is 4.3, and second is 5.21. So, Multiplication = 22.403, addition = 9.51

```python
#  유저한테 두개의 float을 입력받아서
num1 = input("숫자 1 > ")
num2 = input("숫자 2 > ")
```

```python
# 두 수의 곱
# 실수로 형변환 함수 : float(숫자변수)
mul_result = float(num1) * float(num2)

# 합
sum_result = float(num1) + float(num2)
```

```python

aaa = "파이썬"
# 아래 두 문장은 결과 동일
# "안녕 {}".format(aaa)
f"안녕 {aaa}"  # 위 문장의 약식 표현, 더 많이 사용함
```

```python
final_letter = "First is {}, and second is {}. So, Multiplication = {}, addition = {}".format(num1,num2, mul_result,sum_result)
# final_letter = f"First is {num1}, and second is {num2}. So, Multiplication = {mul_result}, addition = {sum_result}"
final_letter
```
