## 항목들을 정렬하기
- 원래의 리스트 자체는 건들지 않고, 새롭게 정렬한 리스트를 반환하는 경우 : sorted()
- 원래의 리스트 자체를 정렬해서 변경하는 경우 : 리스트.sort()

```python
fruits = ['Apple', 'Banana', 'Mango', 'Coconut', 'Orange', 'Lemon']
fruits
```

```python
# 리스트 오름차순
# sorted() : 파이썬 내장함수
sorted(fruits)
```

```python
# 원본에 적용이 안됨
fruits
```

```python
# 리스트 내림차순
sorted(fruits, reverse=True)
```

```python
# 원본이 변경됐을까?
fruits
```

```python
fruits.sort()
```

```python
fruits
```

```python
# 원본이 바뀜
```

```python
fruits
```

```python
# 내림차순
fruits.sort(reverse=True)
```

```python
fruits
```

## 리스트의 특성과 copy() 함수

```python
fruits
```

```python
 fruits2 = fruits
 fruits2
```

```python
fruits2[1] = "Watermelon"
fruits2
```

```python
fruits
```

```python
# 깊은 복사
fruits2 = fruits.copy()
fruits[1] = "Mango"
print(fruits)
print(fruits2)
```

```python
fruits3 = fruits[-3:]
fruits3
```

```python
id(fruits)
```

```python
id(fruits3)
```

```python
 fruits3 = fruits2.copy()
 print(id(fruits2))
 print(id(fruits3))
```
